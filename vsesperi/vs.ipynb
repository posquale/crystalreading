```python
import ROOT

#prendo file simulati
simulazione = ROOT.TFile.Open("simul.root")
simA0 = simulazione.A0
simA = simulazione.A
simB0 = simulazione.B0
simB = simulazione.B

#prendo file misurati
esperimento = ROOT.TFile.Open("espe.root")
espeA0 = esperimento.A0
espeA = esperimento.A
espeB0 = esperimento.B0
espeB = esperimento.B
```

```python

espeB.SetLineWidth(2)
espeB0.SetLineWidth(2)
simB.SetLineWidth(2)
simB0.SetLineWidth(2)
espeB.SetLineColor(1)
espeB0.SetLineColor(1)
simB.SetLineColor(1)
simB0.SetLineColor(1)

simB0.SetLineStyle(1)
simB.SetLineStyle(1)

espeB0.SetFillColorAlpha(16,0.5)
espeB.SetFillColorAlpha(16,0.5)
espeB0.SetLineColorAlpha(16,0)
espeB.SetLineColorAlpha(16,0)

espeB0.SetTitle('Esperimento')
simB0.SetTitle('Simulazione')
espeB.SetTitle('Esperimento')
simB.SetTitle('Simulazione')


canvas0 = ROOT.TCanvas()
canvas0.cd()
pad0=ROOT.TPad()
pad0.Draw()
pad0.cd()
zero=ROOT.THStack("stack0", " ;Carica raccolta (u.a.)")
zero.Add(espeB0)
zero.Add(simB0)

zero.Draw("nostack")
pad0.BuildLegend(0.80,0.85,0.99,0.95,"")
pad0.SetFillColor(0)
canvas0.Draw()

canvas = ROOT.TCanvas()
canvas.cd()
pad=ROOT.TPad()
pad.Draw()
pad.cd()
cut=ROOT.THStack("stack", " ;Carica raccolta (u.a.)")
cut.Add(espeB)
cut.Add(simB)

cut.Draw("nostack")
pad.BuildLegend(0.80,0.85,0.99,0.95,"")
pad.SetFillColor(0)
canvas.Draw()

```

```python
canvas0.SaveAs("Bnocut.png")
canvas.SaveAs("Bcut.png")
```
